```python
import numpy as np
import matplotlib.pyplot as plt
import random

def get_collume(data,index):
    result = [row[index] for row in data]
    return result

def prepare_data(file_name):
    data = np.genfromtxt(file_name,delimiter=',',skip_header=1).tolist()
    N=len(data)
    tv_data=get_collume(data,0)
    radio_data=get_collume(data,1)
    newspaper_data=get_collume(data,2)
    sales_data=get_collume(data,3)
    X=[tv_data,radio_data,newspaper_data]
    Y=sales_data
    #print(X)
    #print(Y)
    return X,Y
# câu 1
X,y = prepare_data('sample_data/advertising.csv')
#list = [sum(X[0][:5]),sum(X[1][:5]),sum(X[2][:5]),sum(y[:5])]
#print(list)

def initialize_params () :
    # w1 = random . gauss ( mu =0.0 , sigma =0.01)
    # w2 = random . gauss ( mu =0.0 , sigma =0.01)
    # w3 = random . gauss ( mu =0.0 , sigma =0.01)
    # b = 0
    w1 , w2 , w3 , b = (0.016992259082509283 , 0.0070783670518262355 ,-0.002307860847821344 , 0)
    return w1 , w2 , w3 , b

def predict ( x1 , x2 , x3 , w1 , w2 , w3 , b ) :
    y_hat = w1 * x1 + w2 * x2 + w3 * x3 + b
    return y_hat

def compute_loss_mse( y , y_hat ):

    # print(f'y: {y}')
    # print(f'y_hat: {y_hat}')
    # print()
    # diff = np.clip(y - y_hat, -1e6, 1e6)
    # loss = diff ** 2
    loss = (y-y_hat) ** 2
    return loss


def compute_gradient_wi ( x , y , y_hat ) :
    gradient = -2 * x * ( y - y_hat )
    return gradient

def compute_gradient_b ( y , y_hat ) :
    gradient = -2 * ( y - y_hat )
    return gradient

def update_weight_wi ( w , dl_dw , lr ) :
    w = w - lr * dl_dw
    return w

def update_weight_b ( b , dl_db , lr ) :
    b = b - lr * dl_db
    return b

def implement_linear_regression_nsamples ( X_data , y_data , epoch_max=50 , lr=1e-5):
    losses = []
    # b = 0.04
    # w = -0.34
    # lr = 0.02
    w1 , w2 , w3 , b = initialize_params ()

    N = len(y_data )
    for epoch in range ( epoch_max ) :
      loss_total = 0.0
      dw1_total = 0.0
      dw2_total = 0.0
      dw3_total = 0.0
      db_total = 0.0
      for i in range ( N ) :
        # get a sample
        x1 = X_data[0][i]
        x2 = X_data[1][i]
        x3 = X_data[2][i]
        y = y_data[i]

        # compute output
        y_hat = predict ( x1 , x2 , x3 , w1 , w2 , w3 , b )

        # compute loss
        loss = compute_loss_mse (y , y_hat )
        # accumulate loss
        loss_total+=loss
        # compute gradient w1 , w2 , w3 , b
        dl_dw1 = compute_gradient_wi ( x1 , y , y_hat )
        dl_dw2 = compute_gradient_wi ( x2 , y , y_hat )
        dl_dw3 = compute_gradient_wi ( x3 , y , y_hat )
        dl_db = compute_gradient_b (y , y_hat )
      # accumulate gradient
        dw1_total += dl_dw1
        dw2_total += dl_dw2
        dw3_total += dl_dw3
        db_total += dl_db
    # update parameters
      w1 = update_weight_wi ( w1 , dw1_total/N , lr )
      w2 = update_weight_wi ( w2 , dw2_total/N , lr )
      w3 = update_weight_wi ( w3 , dw3_total/N , lr )
      b = update_weight_b (b , db_total/N , lr )
      losses.append ( loss_total / N )
    return ( w1 , w2 , w3 ,b , losses )

#print(X,y)
# (w1,w2,w3,b,losses) = implement_linear_regression_nsamples(X , y , epoch_max =1000 ,lr =1e-5)
# print ( losses )
# plt . plot( losses)
# plt . xlabel("# iteration ")
# plt . ylabel(" Loss ")
# plt . show()


#Câu 11
( w1 , w2 , w3 ,b , losses ) = implement_linear_regression_nsamples (X , y ,1000)
print ( w1 , w2 , w3 )
```
